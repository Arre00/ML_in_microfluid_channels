## **Neural**

```python
neural=data
neural
```

```python
X_neural=neural.iloc[:,0:13].values
y_neural=neural.iloc[:, 13:14].values
X_neural
```

```python


from sklearn.model_selection import train_test_split
X_train_nr, X_test_nr, Y_train_nr, Y_test_nr = train_test_split(X_neural, y_neural, test_size=0.2, random_state= 42)
```

```python
from sklearn.linear_model import Ridge
ridge =  Ridge(alpha=0.00)
ridge_lr=ridge.fit(X_train_nr, Y_train_nr)
```

```python
accuracy_score_train_ridge =  ridge_lr.score(X_train_nr,Y_train_nr)
accuracy_score_test_ridge =  ridge_lr.score(X_test_nr,Y_test_nr)
accuracy_score_train_ridge, accuracy_score_test_ridge
```

```python
y_pred_ridge = ridge.predict(X_neural)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_absolute_error(y_neural, y_pred_ridge), ridge.score(X_neural, y_neural)
```

```python
y_graph_neural=[y_neural]
sns.set_style("dark")
plt.figure(figsize=(24,6))
plt.scatter(y_pred_ridge, y_graph_neural, marker='x',s=3)
plt.scatter(y_neural, y_graph_neural, marker='x',s=5)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score, KFold

# define folds
n_folds = 10
kf = KFold(n_folds, shuffle=True)

# define param_grid
l_alpha = [2.**k for k in range(-25, 1)]
param_grid = {'alpha': l_alpha}

# define grid search
lr_m = Ridge()
ridge_alpha_search = GridSearchCV(lr_m, 
                                  param_grid=param_grid, 
                                  cv=kf, 
                                  scoring='neg_mean_absolute_error',
                                  n_jobs=2, 
                                  verbose=1)

_ = ridge_alpha_search.fit(X_neural, y_neural)
```

```python
print("alpha_range: %.2f - %.2f" % (np.array(l_alpha).min(), np.array(l_alpha).max()))
print('alpha = %.2f' % (ridge_alpha_search.best_params_['alpha']))
print('best_cv_mae = %.3f' % (-ridge_alpha_search.best_score_))
sns.set_style("dark")
plt.figure(figsize=(24,6))
plt.xticks(range(len(l_alpha)), l_alpha, rotation=75)

_ = plt.plot( -ridge_alpha_search.cv_results_['mean_test_score'])
```

zero alpha implies linear regression
